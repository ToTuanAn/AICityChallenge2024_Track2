```python
INPUT_PATH = "./data/retrieval/videos"
OUTPUT_PATH = "./data/resized/videos"
CHOICE = ["train", "val", "test"]

for choice in CHOICE:
    for video_path in glob.glob(f"{INPUT_PATH}/{choice}/*.mp4"):
        video_name = video_path.split("/")[-1] 
        input_video_path = f"{INPUT_PATH}/{choice}/{video_name}"
        output_video_path = f"{OUTPUT_PATH}/{choice}/{video_name}"
        print(input_video_path)
        print(output_video_path)
        if not os.path.isfile(output_video_path):
            os.system(f'ffmpeg -i {input_video_path} -vf scale="iw/2:ih/2" {output_video_path}')
```
